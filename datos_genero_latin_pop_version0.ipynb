 Cómo obtener o usar la popularidad
Cada track devuelto incluye un campo llamado popularity dentro del JSON.
Por ejemplo:
for track in datos['tracks']['items']:
    print(track['name'], track['popularity'])
Si quieres ordenar los resultados por popularidad, puedes hacerlo tú mismo en Python:
tracks = datos['tracks']['items']
tracks_sorted = sorted(tracks, key=lambda t: t['popularity'], reverse=True)

for t in tracks_sorted:
    print(f"{t['name']} - Popularidad: {t['popularity']}")
