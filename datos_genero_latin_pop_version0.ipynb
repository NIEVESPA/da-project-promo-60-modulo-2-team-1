Por ejemplo:
for track in datos['tracks']['items']:
    print(track['name'], track['popularity'])
Si quieres ordenar los resultados por popularidad, puedes hacerlo tú mismo en Python:
tracks = datos['tracks']['items']
tracks_sorted = sorted(tracks, key=lambda t: t['popularity'], reverse=True)

for t in tracks_sorted:
    print(f"{t['name']} - Popularidad: {t['popularity']}")

```python

lista_latin_pop = [] #lista vacia antes del bucle

for dato in query['tracks']['items']: #recorremos los elementos
    artista = dato['artists'][0]['name']
    genero = 'latin pop'
    nombre = dato['name']
    tipo = dato['type']
    ano = dato['album']['release_date']
    spotify_id = dato['id']

    informacion_artista= {'Artista': artista, 'Género': 'latin pop', 'Tipo': tipo, 'Nombre': nombre, 'Año': ano,'ID': spotify_id} #creamos un diccionario con la info que queremos recopila de cada truck. nota: genero str

    lista_latin_pop.append(informacion_artista)
    #elimino brake para que me salgan todos los elementos de query

print(lista_latin_pop)
```

```python
# IMPRESIÓN POR FILAS SEPARADAS
print("--- Diccionarios de 'lista_latin_pop' por fila ---")

for diccionario in lista_latin_pop:
    print(diccionario)
```
